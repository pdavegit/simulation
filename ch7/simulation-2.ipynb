```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def Ts(s, lamb, lambt):
    t = s
    # Step 2
    while True:
        U = np.random.uniform()
        t = t - (1.0/lamb)*np.log(U)
        U = np.random.uniform()
        if U <= lambt(t, lamb) / lamb:
            return t
    # Not Reached
```

```python
def lambt(t, lamb):
    return lamb
```

```python
def event_plot(list_t, list_n, title, xlabel, ylabel, color):
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
#ax.plot(list_t, n_t)
    plt.step(list_t, list_n, where='post', label='post')
    plt.plot(list_t, list_n,  color, alpha=0.5)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
```

```python
def service_time():
    # return np.random.uniform()*20.0/60.0
    return gen_var(1)[0]/60.0
```

```python
def single_server(lamb):
    Y = 1
    A = []
    D = []
    N = []
    Event_Time = []
    Y_Array = []
    t, NA, ND = 0, 0, 0
    T = 100
    n = 0
    T0 = Ts(0, lamb, lambt)
    tA = T0
    tD = np.Inf
    while True:
        # Case 1
        if tA <= tD and tA <= T:
            # Arrival
            t = tA
            NA = NA + 1
            n = n + 1
            Tt = Ts(t, lamb, lambt)
            tA = Tt
            if n == 1:
                Y = service_time()
                tD = t + Y
                Y_Array.append(Y)
            A.append(t)
            N.append(n)
            Event_Time.append(t)
        elif tD < tA and tD <= T:
            # Departure
        # Case 2
            t = tD
            n = n - 1
            ND = ND + 1
            if n == 0:
                tD = np.Inf
            else:
                Y = service_time()
                tD = t + Y
                Y_Array.append(Y)
            D.append(t)
            N.append(n)
            Event_Time.append(t)
        elif np.minimum(tA, tD) > T and n > 0:
            # Departure after door closes
            # Case 3
            t = tD
            n = n - 1
            ND = ND + 1
            if n > 0:
                Y = service_time()
                tD = t + Y
                Y_Array.append(Y)
            D.append(t)
            N.append(n)
            Event_Time.append(t)
        elif np.minimum(tA, tD) > T and n == 0:
            Tp = np.maximum(t - T, 0)
            # return
            return (A, D, N, Event_Time, Tp, Y_Array)
        # Case 4
        else:
            raise "Not Reached"
```

```python
total_arrivals = []
for i in range(1):
    (A, D, N, Event_Time, Tp, Y_Array) = single_server(35)
    if len(Event_Time) != len(N):
        raise "Error - Expected both arrays to be same length"
    event_plot(A, np.arange(len(A)) + 1, "Arrivals", "t", "A", 'C1o')
    event_plot(D, np.arange(len(A)) + 1, "Departures", "t", "D", 'C2o')           
    event_plot(Event_Time, N, "State", "t", "n", 'C3o')
    fig = plt.figure()
    plt.hist(Y_Array, bins=100)
    print(Tp)
    total_arrivals.append(len(A))
print(np.mean(total_arrivals))
```
